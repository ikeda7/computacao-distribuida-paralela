```python
import multiprocessing as mp
import os
import numpy as np
```

## mostre o número de processadores/núcleos

```python
print("Números de processadores/cores: ", mp.cpu_count())
```

## crie uma fila com 10 números inteiros

```python
fila = mp.Queue()
inteiros = np.arange(1, 11)
inteiros = inteiros.tolist()
_ = [fila.put(i) for i in inteiros]
```

## calcule e mostre o triplo de um número da fila

```python
def get_triple(num: int):
    return num * 3

print(get_triple(inteiros[0]))
```

## crie um pool de processos para executar a função de cálculo do triplo

### teste com o mesmo número de núcleos

```python
pool = mp.Pool(mp.cpu_count())
print(pool.map(get_triple, inteiros))
pool.close()
pool.join()
```

### e com diferente

```python
extra_processes = 2
pool = mp.Pool(mp.cpu_count() + extra_processes)
print(pool.map(get_triple, inteiros))
pool.close()
pool.join()
```
